```python
from litellm import completion
from typing import List, Dict
import os
import json

api_key = "key_here"
os.environ["GROQ_API_KEY"] = "key_here"
api_key = os.getenv("GROQ_API_KEY")

def generate_response(messages: List[Dict]) -> str:
    response = completion(
        model="groq/llama-3.1-8b-instant",
        messages=messages,
        temperature=0,
        max_tokens=200
    )
    return response.choices[0].message.content

code_spec = {
    'name': 'program',
    'description': 'You are a software programmer & use functional programming',
    'params': {
        'd': 'program for integers'
    },
}

messages = [
    {"role": "system",
     "content": "You are an expert software engineer that writes python program."},
    {"role": "user", "content": f"Please implement: {json.dumps(code_spec)}"}
]

response = generate_response(messages)
print(response)
```
